```python
news_dates=[]
def find_news_dates(stockname):
    if(stockname.lower()[0]=='s'):
        os.chdir('E:\\SEM VIII\\Final Year Project\\News_Feed\\SUN PHARMA')
    elif(stockname.lower()[0]=='p'):
        os.chdir('E:\\SEM VIII\\Final Year Project\\News_Feed\\PFIZER')
    elif(stockname.lower()[0]=='d'):
        os.chdir('E:\\SEM VIII\\Final Year Project\\News_Feed\\DR REDDY')
    dates=os.listdir()
    for i in dates:
        news_dates.append(i.split('.')[0])
    return news_dates
```

```python
intersecting_dates=[]
def match_news_dates(news_dates,data):
    stock_data=data['timestamp'].tolist()
    dates_set=set(stock_data).intersection(news_dates)
    for i in dates_set:
        intersecting_dates.append(i)
    return intersecting_dates
```

```python
data_for_analysis={}
def load_and_retrieve_json_data(stockname,relevant_dates,data_for_analysis):
    if(stockname.lower()[0]=='s'):
        os.chdir('E:\\SEM VIII\\Final Year Project\\News_Feed\\SUN PHARMA')
    elif(stockname.lower()[0]=='p'):
        os.chdir('E:\\SEM VIII\\Final Year Project\\News_Feed\\PFIZER')
    elif(stockname.lower()[0]=='d'):
        os.chdir('E:\\SEM VIII\\Final Year Project\\News_Feed\\DR REDDY')
    for i in range(len(relevant_dates['timestamp'])):
        with open(relevant_dates['timestamp'][i]+".json") as json_data:
            news_json=json.load(json_data)
            data_for_analysis[relevant_dates['timestamp'][i]] = {'titles': [], 'descriptions': [], 'contents': []}
            for j in news_json['articles']:
                data_for_analysis[relevant_dates['timestamp'][i]]['titles'].append(j['title'])
                data_for_analysis[relevant_dates['timestamp'][i]]['descriptions'].append(j['description'])
                data_for_analysis[relevant_dates['timestamp'][i]]['contents'].append(j['content'])
    return data_for_analysis
```

```python
def predict_sentiment(text):
    value=TextBlob(text).sentiment.polarity
    return value
```

```python
sentiment_dict={}
def assign_sentiment_values(data_for_analysis):
    for i in data_for_analysis.keys():
        sentiment_dict[i]={'title_values':[],'description_values':[],'content_values':[]}
        for j in data_for_analysis[i]['titles']:
            sentiment_dict[i]['title_values'].append(predict_sentiment(j))
        for j in data_for_analysis[i]['descriptions']:
            sentiment_dict[i]['description_values'].append(predict_sentiment(j))
        for j in data_for_analysis[i]['contents']:
            if type(j)==str:
                sentiment_dict[i]['content_values'].append(predict_sentiment(j))
    return sentiment_dict
```

```python
assign_sentiment_values(data_for_analysis)
```

```python
infer_sentiment_dict={}
average_title_value=0
average_description_value=0
average_content_value=0
def infer_sentiment(sentiment_dict):
    global average_title_value, average_description_value, average_content_value
    for i in sentiment_dict.keys():
        infer_sentiment_dict[i]={'title_value':'','description_value':'','content_value':''}
        for j in sentiment_dict[i]['title_values']:
            average_title_value=average_title_value+j
        average_title_value=average_title_value/len(sentiment_dict[i]['title_values'])
        infer_sentiment_dict[i]['title_value']=average_title_value
        for j in sentiment_dict[i]['description_values']:
            average_description_value=average_description_value+j
        average_description_value=average_description_value/len(sentiment_dict[i]['description_values'])
        infer_sentiment_dict[i]['description_value']=average_description_value
        for j in sentiment_dict[i]['content_values']:
            average_content_value=average_content_value+j
        average_content_value=average_content_value/len(sentiment_dict[i]['content_values'])
        infer_sentiment_dict[i]['content_value']=average_content_value    
    return infer_sentiment_dict
```

```python
# 1-UP & 0-DOWN
def assign_label(data):
    initial_value=0
    data=data.assign(labels=np.where(data['close']-data['open']>0,1,initial_value))
    return data["labels"]
```

```python
temporary_list=[]
description_list=[]
content_list=[]
def insert_sentimental_values_in_list(relevant_dates,infer_sentiment_dict):
    for i in relevant_dates.index.values:
        temporary_list.append(infer_sentiment_dict[i]["title_value"])
    for i in relevant_dates.index.values:
        description_list.append(infer_sentiment_dict[i]["description_value"])
    for i in relevant_dates.index.values:
        content_list.append(infer_sentiment_dict[i]["content_value"])
    return temporary_list,description_list,content_list
```
